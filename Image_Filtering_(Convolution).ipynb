```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

### Load the image using OpenCV

```python
im = cv2.imread('/content/lenna.png') #change the path as needed
```

```python
im.shape
```

### Convert the RGB image into a grayscale image

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

### Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
h_kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

### Do the filtering

```python
v_image = cv2.filter2D(im, -1, v_kernel)
h_image = cv2.filter2D(im, -1, h_kernel)
```

### Display results

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')
```

```python
im = cv2.imread('/content/lenna.png')
```

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
sobel_v_kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])
```

```python
sobel_h_kernel = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
])
```

```python
sobel_v_image = cv2.filter2D(im, -1, sobel_v_kernel)
sobel_h_image = cv2.filter2D(im, -1, sobel_h_kernel)
```

```python
cv2.imwrite('/content/sobel_vertical.png', sobel_v_image)
cv2.imwrite('/content/sobel_horizontal.png', sobel_h_image)
```

```python
# Visualize the results
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(sobel_v_image, cmap='gray')
plt.title('Sobel Vertical Edges')

plt.figure()
plt.imshow(sobel_h_image, cmap='gray')
plt.title('Sobel Horizontal Edges')

plt.show()
```
